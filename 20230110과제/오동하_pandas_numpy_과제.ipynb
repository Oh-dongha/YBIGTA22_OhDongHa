#### birthDate 컬럼으로 age 컬럼 추가하기

##### 1) birthDate 컬럼의 자료형을 object에서 datetime으로 전환

```python
data['birthDate'] = pd.to_datetime(data['birthDate'])
print(data["birthDate"].dtype)
```

##### 2) age 컬럼 생성, apply 및 lambda 함수를 써서 나이 계산

```python
data['age'] = data['birthDate'].apply(lambda x: datetime.today()-x)
data['age'] = data['age'] / np.timedelta64(1, 'Y')
print(data.head(30))
```

##### 3) age 결측치를 선수 평균 나이로 대체

```python
datamean=data.fillna(data.mean()['age':"age"])
print(datamean.tail(30))
```

#### height 컬럼(feet-inch 형식)으로 heightCm 컬럼 추가하기

##### 1) heightCm 컬럼 생성, apply 및 lambda 함수를 사용해서 height 컬럼의 feet과 inch를 cm로 변환 ('-'를 기준으로 split해서 각각을 cm로 변환해 더하기)



```python
data['height'] = data['height'].astype('string')
data['heightCm'] = data['height'].apply(lambda x: int(x[0])*(30.48)+int(x[2:])*(2.54))

print(data.tail(30))
```
